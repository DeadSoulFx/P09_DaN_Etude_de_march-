##### Calcul des totaux par pays

```python
# Le but est de calculer les différents totaux par pays puis de les ajouter au dataset de dispo alim
# On gropuby sur zone et item et on .sum(), cela nous donne le total pour chaque item par pays
dftest = data_dispo_alim.groupby(['zone', 'item']).agg({'value' : 'sum'}).reset_index()
dftest
```

```python
Production + Importation - Exportation + Variation
= Dispo =
Aliments animaux + Autres utilisations + Nourriture + Pertes + Semence + Traitement
```

```python
# A déplacer plus loin, c'était un test pour le concat
pivot_test = pd.concat([data_dispo_alim, dftest]).fillna({'product' : 'total_zone'})
pivot_test.head(20)
```

```python
pvt = pivot_test.pivot_table(index='zone', values='value', columns=['item', 'product']).reset_index()
pvt
```

```python
# On créé un nouveau dataframe avec les données qui nous intéressent, ce sera beaucoup plus propre que de supprimer des colonnes ou des lignes au préalable
df = pd.DataFrame()
df['pays'] = pvt['zone']
df['dispo_chicken_kcal'] = pvt['Disponibilité alimentaire (Kcal/personne/jour)']['Viande de Volailles']
df['dispo_total_kcal'] = pvt['Disponibilité alimentaire (Kcal/personne/jour)']['total_zone']
df
```

##### Dispo alim kcal Totale Pays

```python
# Calcul et ajout de la dispo kCal globale par pays, on prend le global de tous les produits
# On prend uniquementles lignes de dispo alim en kCal
temp_alim_tot = data_dispo_alim.loc[data_dispo_alim['item'] == 'Disponibilité alimentaire (Kcal/personne/jour)']
temp_alim_tot

# On sum la  dispo kCal par pays pour avoir les kCal par pays de dispo
temp_alim_tot_gb_country = temp_alim_tot.groupby(['zone']).agg({'value' : 'sum'}).reset_index()
temp_alim_tot_gb_country

# Test si ça matche, non on concat au fur et à mesure pour ajouter nos items à chaque fois
# On ajoute la dispo totale calculé à la suite du df de base et on remplace l'item par ce qu'il est, on ne touche pas aux autres colonnes pour pouvoir concat avec la population
# Sur les lignes rajoutées qui on un NaN, on indique des infos pour savoir la correspondance sur des colonnes (à affiner pour la suite)
# Pour les Item et Unity on reprend la même trame que le df
pd.concat([data_dispo_alim, temp_alim_tot_gb_country]).fillna({'item' : 'Disponibilité alimentaire Totale (Kcal/personne/jour)', 'unity' : 'Kcal/personne/jour'})
```
